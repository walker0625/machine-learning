# LDA

```python
import pandas as pd

data = pd.read_csv("data/appreply2.csv")
```

```python
from konlpy.tag import Okt
import re

okt = Okt()

review_list = []

for review in data['text'].tolist():

    rv = re.sub(r"[^0-9a-zA-Z가-힣+-]", "", review)
    
    result = okt.morphs(rv)

    review_list.append(' '.join(result)) # vectorize를 위한 공백(' ') 구분 형태로 review 전처리
    
for r in review_list:
    print(r)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(
    max_df = 0.1,               # 전체 단어 등장 비율이 p 이상인 것만 사용할 것.
    min_df = 2,                 # 단어가 적어도 n개 이상인 것만 사용할 것.
    max_features = 1000,        # 상위 항목 n개만 사용할 것.
    ngram_range= (1,2)          # 단어(ngram)의 조합
)
```

```python
feature_vec = vectorizer.fit_transform(review_list)
feature_vec.shape
```

```python
feature_names = vectorizer.get_feature_names_out()
print(feature_names)
```

```python
from sklearn.decomposition import LatentDirichletAllocation

# 토픽(행) 별로 중요한 단어(열)를 추출
lda = LatentDirichletAllocation(n_components=5)
lda.fit(feature_vec)
```

```python
print(lda.components_.shape)
print(lda.components_)
```

```python
import numpy as np 

test = np.array([30, 10, 20])
print(test.argsort())
```

```python
for idx, topic in enumerate(lda.components_):
    print(f"# Topic {idx+1}:")
    topic_word_idx = topic.argsort()[::-1]
    top_idxs = topic_word_idx[:20]
    topics = [feature_names[idx] for idx in top_idxs]
    
    print(" ".join(topics)) # 각 토픽 문서에서 공통 주제와 관련있는 키워드 상위 20개
```

```python
import pyLDAvis.lda_model 

pyLDAvis.enable_notebook()
vis = pyLDAvis.lda_model.prepare(lda, feature_vec, vectorizer)
pyLDAvis.display(vis)
```

```python
# 비지도 학습(분류가 정해지지 않은 상황에서 라벨링하여 접근 가능)
# 어떤 문장이 어떤 분류에 들어갈 확률
sentence_topic = lda.transform(feature_vec)
print(sentence_topic.shape)
print(sentence_topic)
```

```python
sent_list = []

for n in range(sentence_topic.shape[0]):
    topic_idx = sentence_topic[n].argmax()
    topic_prob = sentence_topic[n].max()

    sent_list.append([n, topic_idx, topic_prob])

topic_per_data = pd.DataFrame(sent_list, columns=["no", "topic_no", "prob"])
topic_per_data["review"] = data["text"]
topic_per_data
```

```python
topic_per_data['topic_no'].value_counts()
```

```python
for topic in range(topic_per_data["topic_no"].nunique()):
    temp_data = topic_per_data.loc[topic_per_data["topic_no"]==topic, :]
    print(temp_data.sort_values(by=["prob"], ascending=False).iloc[:5,-1])
    print("="*100)
```

```python
# 지도 학습

data['reaction'] = np.where(data['score'] >= 4, 1, 0)
data.head()
```

```python
# 너무 긴 리뷰는 분석의 유효성을 해침(1. 단어 종류가 너무 많아짐 2. 0이 많아짐) - 히스토그램으로 확인하여 수정/삭제 필요

data['length'] = data['text'].str.len()
data.head()
```

```python
import matplotlib.pyplot as plt 
import koreanize_matplotlib 

plt.figure(figsize=(5,5))
data["length"].plot(kind="hist")
plt.title("문장의 길이 히스토그램")
plt.show()
```

```python
from konlpy.tag import Okt
import re

okt = Okt()

review_list = []

for review in data['text'].tolist():

    rv = re.sub(r"[^0-9a-zA-Z가-힣\s]", "", review)
    
    result = okt.morphs(rv)

    review_list.append(' '.join(result)) # vectorize를 위한 공백(' ') 구분 형태로 review 전처리
    
for r in review_list:
    print(r)
```

```python
# TfidfVectorizer

from sklearn.feature_extraction.text import TfidfVectorizer 

vectorizer = TfidfVectorizer()
tfidf_vec = vectorizer.fit_transform(review_list)
print(vectorizer.vocabulary_)
print(tfidf_vec.toarray())
```

```python
# 학습을 위한 data 준비

# Train, Test 데이터셋 나누기
from sklearn.model_selection import train_test_split 

X = tfidf_vec
y = data["reaction"]

# random_state을 통해 재실행시에 동일한 결과를 보장(seed)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=22)

print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# 데이터 학습

from sklearn.linear_model import LogisticRegression 

lr_model = LogisticRegression(random_state = 0)
lr_model.fit(X_train, y_train)
```

```python
y_pred = lr_model.predict(X_test)
print(y_pred)
```

```python
# 성능 평가

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score 

print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print(f"Precision: {precision_score(y_test, y_pred)}")
print(f"Recall: {recall_score(y_test, y_pred)}")
print(f"F1-Score: {f1_score(y_test, y_pred)}")
```

```python
import pandas as pd

df = pd.read_csv("data/appreply2.csv")
```

```python
from konlpy.tag import Okt
import re

okt = Okt()

review_list = []

for review in df['text'].tolist():

    rv = re.sub(r"[^0-9a-zA-Z가-힣+-]", "", review)
    
    result = okt.morphs(rv)

    review_list.append(' '.join(result))
    
for r in review_list:
    print(r)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

print(review_list[0])

vectorizer = CountVectorizer(
    max_df=0.1,
    min_df=2,
    max_features=1000,
    ngram_range=(1,2)
)
```

# 성능 평가

```python
# 정확도 - 클래스 불균형 상황에서 신뢰도 저하(희귀암)
# 정확도를 중시하는 문제 상황 -> Data 불균형이 적은 상황 -> Data보다는 Model이 중요
from sklearn.metrics import accuracy_score

y_true = [1, 0, 1, 1, 0, 1]
y_pred = [1, 0, 1, 0, 0, 1]

accuracy_score(y_true, y_pred)
```

```python
# 정밀도 - 결과에 따른 효과가 치명적이지 않은 경우 유용(ex : 스팸메일 판단)
# 모델이 스팸이라고 한 것 중에, 실제 스팸인 경우
from sklearn.metrics import precision_score

precision_score(y_true, y_pred)
```

```python
# 리콜 - 정확도가 매우 중요한 분야(암 판정) - 오류가 발생하더라도 맞다고 하는게 안정적
# 실제 암인 경우를, 모델이 암이라고 하는 경우
from sklearn.metrics import recall_score

recall_score(y_true, y_pred)
```

```python
# F1 Score - 정밀도와 리콜의 반비례성을 상호보완하기 위해 생성한 지표
# F1 Score가 중요한 상황은 Data가 불균형한 상황(의료) -> 전처리 중요
# Data가 균형 상태이면 정확도를 중시하는 것이 나을 수 있음

from sklearn.metrics import f1_score

f1_score(y_true, y_pred)
```

```python
# Confusion Matrix - 분류 모델에만 사용
from sklearn.metrics import confusion_matrix

confusion_matrix(y_true, y_pred)
```

```python
# Precision, Recall, F1, Support 한번에!(macro avg 확인)
from sklearn.metrics import classification_report

print(classification_report(y_true, y_pred))
```
